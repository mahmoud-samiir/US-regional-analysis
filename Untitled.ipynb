# IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

# READ DATA

```python
data = pd.read_csv('US_Regional_Sales_Data.csv')
data.head()
```

# INITIAL PLAN FOR DATA EXPLORATION

```python
data.info()
```

```python
data['ProcuredDate']=pd.to_datetime(data['ProcuredDate'])
data['OrderDate']=pd.to_datetime(data['OrderDate'])
data['ShipDate']=pd.to_datetime(data['ShipDate'])
data['DeliveryDate']=pd.to_datetime(data['DeliveryDate'])
# Remove commas and convert to float
data['Unit Cost'] = data['Unit Cost'].str.replace(',', '').astype(float)
data['Unit Price'] = data['Unit Price'].str.replace(',', '').astype(float)
```

```python
print(data['OrderNumber'].nunique())
```

```python
print(((data['ProcuredDate'] <= data['OrderDate'])&
      (data['OrderDate']<=data['ShipDate'])&
      (data['ShipDate']<=data['DeliveryDate'])).sum())
```

```python
cleaned_data=data[(data['ProcuredDate'] <= data['OrderDate'])&
      (data['OrderDate']<=data['ShipDate'])&
      (data['ShipDate']<=data['DeliveryDate'])]
```

```python
cleaned_data.head()
```

# DATA CLEANING AND FEATURE ENGINEERING 

```python
print((cleaned_data['Unit Price']>=cleaned_data['Unit Cost']).sum())
```

```python
cleaned_data.duplicated().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=cleaned_data)
plt.title('Box Plot for Numerical Columns')
plt.xticks(rotation=45)
plt.show()
```

```python
columns_to_clean = ['Unit Cost', 'Discount Applied']
for column in columns_to_clean:
    Q1 = cleaned_data[column].quantile(0.25)
    Q3 = cleaned_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_data = cleaned_data[(cleaned_data[column] >= lower_bound) & (cleaned_data[column] <= upper_bound)]
```

```python
cleaned_data['Order_to_Delivery_Duration'] = (cleaned_data['OrderDate'] - cleaned_data['DeliveryDate']).dt.days
```

```python
cleaned_data['Profit_Margin'] = cleaned_data['Unit Price']-cleaned_data['Unit Cost']
```

# EXPLORATORY DATA ANALYSIS

```python
cleaned_data['Sales Channel'].value_counts().plot(kind='bar', title='Sales Channel Distribution')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = cleaned_data.groupby('_ProductID')['Unit Price'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='barh', color='orange', title='Top 5 Products by Revenue')
plt.xlabel('Revenue')
plt.show()
```

```python
sns.scatterplot(x='Discount Applied', y='Order Quantity', data=cleaned_data, hue='Sales Channel', palette='viridis')
plt.title('Discount vs Order Quantity')
plt.show()
```

```python
cleaned_data['WarehouseCode'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='Pastel1', title='Warehouse Utilization')
plt.ylabel('')
plt.show()
```
